## Desafio 1


Tendo como referência a imagem abaixo (no repositório, “imagens/nota-fiscal.png”), realize alguma transformação de tal forma que fique somente o boleto na imagem, sem o background, ajustando sua perspectiva.
Tal operação é similar utilizado em aplicações de escaneamento de documentos.


![alt text](imagens/nota-fiscal.png)

```python
import cv2
print(cv2.__version__)

import numpy as np

from matplotlib import pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline

plt.style.use('seaborn')
```

```python
image = cv2.imread("imagens/nota-fiscal.png")

cv2.imshow("Nota", image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image_plt = mpimg.imread("imagens/nota-fiscal.png")
plt.imshow(image_plt)
plt.show()
```

```python
height, width = image.shape[:2]
print(height, width)

image_marker = image.copy()

pts1_marker = np.float32([[120,20],[400,20],[90,620],[460,620]])
pts2_strech = np.float32([[0,0],[650,0],[0,650],[550,650]])

color = (0, 0, 255)

cv2.circle(image_marker, (120,20), 5, color, 3)
cv2.circle(image_marker, (400,20), 5, color, 3)
cv2.circle(image_marker, (90,620), 5, color, 3)
cv2.circle(image_marker, (460,620), 5, color, 3)

cv2.imwrite("nota-fiscal-proc.png", image_marker)

image_plt = mpimg.imread("nota-fiscal-proc.png")
plt.imshow(image_plt)
plt.show()

M = cv2.getPerspectiveTransform(pts1_marker, pts2_strech)
dst = cv2.warpPerspective(image,M,(550,650))

cv2.imwrite("nota-fisca-persp.png", dst)

image_plt = mpimg.imread("nota-fisca-persp.png")
plt.imshow(image_plt)
plt.show()
```

```python
cv2.imshow("Nota", dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
